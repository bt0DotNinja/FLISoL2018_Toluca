## Simulación MonteCarlo

El método de Monte Carlo es un método no determinista o estadístico numérico, usado para aproximar expresiones matemáticas complejas y costosas de evaluar con exactitud

La invención del método de Monte Carlo se asigna a Stanislaw Ulam y a John von Neumann. Ulam ha explicado cómo se le ocurrió la idea mientras jugaba un solitario durante una enfermedad.

## Numeros Pseudo Aleatorios

Usaremos un modulo estandar para generar numeros Pseudo Aleatorios

```python
import random

# probemos

random.random()
```

Escribamos nuestra primera función: 

```python
def LanzarChispitas(numChispitas):
    inCircle=0
    for chispitas in range(numChispitas):
        x=random.random()
        y=random.random()
        if (x*x + y*y)**0.5 <= 1.0:
            inCircle += 1
    return 4*(inCircle/float(numChispitas))
```

```python
pi = LanzarChispitas(10000000)
```
